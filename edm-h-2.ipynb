## Descriptive statistical Analysis

```python
low_level_students = df[df['StageID']=='lowerlevel']
low_level_students
```

```python
df['NationalITy'].describe()
```

```python
df.groupby('NationalITy').count()
```

```python
 nationality = df.groupby('NationalITy').count().iloc[:,1].sort_values()
 nationality
```

```python
countries = list(nationality.index)
print(countries)
students = list(nationality.values)
print(students)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.bar(countries,students)
plt.show()
```

```python
df_male = df.loc[df['gender'] == 'M']
print(len(df_male))

df_female = df.loc[df['gender'] == 'F']
print(len(df_female))
```

```python
labels = ['Male','Female']
sizes= [len(df_male), len(df_female)]
fig1, ax1 = plt.subplots()

fig1.set_size_inches(8,8)
ax1.pie(sizes,labels=labels, autopct='%1.1f%%')
ax1.set_title('Male vs Female')
ax1.axis('equal')
plt.show()
```

```python
df_female
```

```python
print('Female students ')
female_nationality = df_female.groupby('NationalITy').count().iloc[:,1]
print(female_nationality)
female_students = list(female_nationality.values)
female_students_countries = list( female_nationality.index)
print(female_students_countries)
print(female_students)

print('Male student')
male_nationality = df_male.groupby('NationalITy').count().iloc[:,1]
print(male_nationality)
male_students = list(male_nationality.values)
male_students_countries = list( male_nationality.index)
print(male_students_countries)
print(male_students)
print(len(female_students_countries) == len(male_students_countries))
```

```python
fig = plt.figure()
data = [female_students, male_students]
ax = fig.add_axes([0,0,2,2])

ax.bar(male_students_countries,male_students)
ax.bar(female_students_countries, female_students)

ax.legend(labels=['Male', 'Female'])
ax.set_xlabel('Countries')
ax.set_ylabel('Number of students')
ax.set_title('Number of male and female students by country')

plt.show()
```
